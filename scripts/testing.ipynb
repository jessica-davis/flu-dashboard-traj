```python
import pandas as pd
import os
import epiweeks
```

```python
files = os.listdir("../raw/multistrain_retrospective_trajectories/")
#files = ['trajectories_aggregated_202608.csv']#, 'trajectories_aggregated_202607.csv']
all_data = []
for file in files:
    if file[0] == ".":
        continue

    df = pd.read_csv("../raw/multistrain_retrospective_trajectories/%s" % file)
    df['date'] = pd.to_datetime(df['date'])
    file_horizon0 = file.split("_")[-1].split(".")[0]
    df['reference_date'] = pd.to_datetime(epiweeks.Week.fromstring(file_horizon0).enddate())    
    df_filter = df[df.date>=df.reference_date]
    df_filter['horizon'] = (df_filter['date'] - df_filter['reference_date']).dt.days//7
    df_filter.rename(columns={"location":"location_name_epydemix"})
    df_filter['epiweek'] = df_filter['date'].apply(lambda x: epiweeks.Week.fromdate(x).cdcformat())

    df_filter = df_filter[df_filter.horizon<=3]

    df_filter.to_parquet("../raw/multistrain_retrospective_trajectories/%s" % "epystrain_trajectories_%s.parquet" % pd.to_datetime(epiweeks.Week.fromstring(file_horizon0).enddate()).strftime("%Y-%m-%d"), index = False)
    #all_data.append(df_filter) 
    
#df_final = pd.concat(all_data)
```
